# 1. StructType (Nested Object)

### Concept

`StructType` represents a **nested object / record** inside a column.

Think of it like a **JSON object**.

Example JSON:

```json
{
  "id": 1,
  "name": "John",
  "address": {
    "city": "Chennai",
    "pincode": 600001
  }
}
```

Here **address** is a **struct**.

---

### PySpark Representation

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),

    StructField(
        "address",
        StructType([
            StructField("city", StringType(), True),
            StructField("pincode", IntegerType(), True)
        ]),
        True
    )
])
```

---

### Access Struct Fields

```python
df.select("address.city").show()

# OR

df.select(df.address.city).show()
```

---

# 2. StructField

### Concept

`StructField` defines **one column inside a schema**.

It has 3 main properties:

```
StructField(
    name,
    dataType,
    nullable
)
```

Example

```python
StructField("id", IntegerType(), True)
```

Meaning:

| Parameter | Meaning             |
| --------- | ------------------- |
| name      | column name         |
| dataType  | column data type    |
| nullable  | True = null allowed |

---

### Example Schema

```python
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True)
])
```

---

# 3. ArrayType (List / Collection)

### Concept

`ArrayType` represents a **list of values**.

Example JSON:

```json
{
  "id": 1,
  "skills": ["python", "spark", "sql"]
}
```

Here **skills** is an **array**.

---

### PySpark Schema

```python
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("skills", ArrayType(StringType()), True)
])
```

Meaning:

```
skills -> array<string>
```

---

### Access Array Values

Example

```python
df.select("skills").show()
```

Get first element:

```python
df.select(df.skills[0]).show()
```

Flatten array using explode:

```python
from pyspark.sql.functions import explode

df.select(explode("skills")).show()
```

---

# 4. MapType (Key Value Pair)

### Concept

`MapType` represents **dictionary-like structure**.

Example JSON

```json
{
  "id": 1,
  "marks": {
    "math": 90,
    "english": 85
  }
}
```

Here **marks** is a **map**.

---

### PySpark Schema

```python
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("marks", MapType(StringType(), IntegerType()), True)
])
```

Meaning:

```
marks -> map<string, int>
```

Key = string
Value = integer

---

### Access Map Values

```python
df.select("marks['math']").show()
```

````python


from pyspark.sql.types import StructType,StructField, StringType, IntegerType, ArrayType, MapType
from pyspark.sql.functions import *

df = spark.createDataFrame([(jsonStr,)],schema=["jsonStr"])
#df.show()



schema = StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType()),
    StructField("age", IntegerType()),
 StructField("address",StructType([
     StructField("city",StringType()),
     StructField("state",StringType()),
     StructField("pincode",IntegerType())
 ])),

 StructField("skills",dataType=ArrayType(StringType())), 
 StructField("projects",dataType=ArrayType(
     StructType([
         StructField("project_name",StringType()),
         StructField("duration",IntegerType())
     ])
 )),
 StructField("ratings",dataType=MapType(StringType(),IntegerType()))
 ])


df=df.withColumn("value",from_json("jsonStr",ArrayType(schema)))



df.select(explode("value").alias("value")).select("value.*").show(truncate=False)
````
