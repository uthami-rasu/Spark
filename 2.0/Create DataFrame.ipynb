# 1. With Default Schema

```python
data = ((1,"razz",28),(2,"ram","28"))



df = spark.createDataFrame(data,["id","name","age"])


df.show()
```

# 2. With Explicit Schema

```python
from pyspark.sql import types as T 
from pyspark.sql import functions as F 



schema = T.StructType([
    T.StructField(
        name="id", dataType=T.IntegerType(),nullable=False
    ),
     T.StructField(
        name="name", dataType=T.StringType(),nullable=False
    ), T.StructField(
        name="age", dataType=T.IntegerType(),nullable=False
    ),
])



data = [
    (1,"razz",28),(2,"ram",28)
]


df = spark.createDataFrame(data=data,schema=schema)

df.show() 
```

# 3. Using a List of Dictionaries

```python
data = [
    {
        "name" : 'Razz', 
        "age" : 28, 
        "id" : 1
    },
    {
        "name" : 'Ram', 
        "age" : 28, 
        "id" : 2
    }
]




schema = "id int, name string, age int" 

df = spark.createDataFrame(data=data,schema=schema)

df.show()
```
