```python
# write csv as a delta table 



csvFile = r'/Volumes/learn-spark/default/learn-spark/AVCPL-1_packing_list_20260610_063118_delimterchanged.csv/' 

df = spark.read.format("csv").option("header",True).option("delimiter","|").load(csvFile)

# Rename columns to remove spaces (Delta tables don't allow spaces in column names)
df = df.toDF("Company_Code", "Inventory_Code", "Inventory_Description", "Price", "Quantity")

df = df.dropna(how="any")



df.write.mode("overwrite").saveAsTable("packing_list")
```
