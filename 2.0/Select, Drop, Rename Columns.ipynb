```python
df = spark.read.format("csv").options(**{
    "header":"true","delimiter":"|","inferSchema":"true"
}).load("/Volumes/learn-spark/default/learn-spark/AVCPL-1_packing_list_20260610_063118_delimterchanged.csv/")



df = df.dropna('any')
df.show()
```

```python
# select columns 



df.select('Company Code').show() 
```

```python
# select mulotiple columns 


df.select("Company Code","Inventory Code").show()
```

```python
# Drop columns 


df.drop("Company Code").show()
# Rename columns 


df.withColumnRenamed("Company Code","Company_Code").show()
# Add columns 

```

```python
# Adding Columns 

from pyspark.sql.functions import * 

df.withColumn("total_amount",round(col("price") * col("Quantity"),2)).show()


df.withColumn("isBulk",when(col("price")  > 1000000, "yes").otherwise("no")).show()
```
