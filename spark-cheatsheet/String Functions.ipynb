# 🔹 PySpark String Functions Cheat Sheet

```python
from pyspark.sql import functions as F
```

---

## 🔸 1. Case Conversion

```python
F.upper(col)        # → UPPERCASE
F.lower(col)        # → lowercase
F.initcap(col)      # → Capitalize Each Word
```

---

## 🔸 2. Trimming / Cleaning

```python
F.trim(col)         # remove leading & trailing spaces
F.ltrim(col)        # remove leading spaces
F.rtrim(col)        # remove trailing spaces
```

---

## 🔸 3. Length & Substring

```python
F.length(col)               # string length
F.substring(col, 1, 3)      # start pos (1-based), length
F.substr(col, 1, 3)         # alias
```

---

## 🔸 4. Concatenation

```python
F.concat(col1, col2)                     # combine columns
F.concat_ws("-", col1, col2)             # with separator
```

---

## 🔸 5. Replace / Translate

```python
F.replace(col, "old", "new")             # (Spark 3.5+)
F.regexp_replace(col, "pattern", "new")  # regex replace
F.translate(col, "abc", "xyz")           # char mapping
```

---

## 🔸 6. Search / Match

```python
F.instr(col, "sub")              # position of substring
F.locate("sub", col)             # same as instr
F.contains(col, "text")          # boolean
F.startswith(col, "text")
F.endswith(col, "text")
```

---

## 🔸 7. Regex Functions

```python
F.regexp_extract(col, r"(\d+)", 1)   # extract match
F.regexp_replace(col, r"\d+", "")    # remove digits
```

---

## 🔸 8. Split & Join

```python
F.split(col, ",")                # → array
F.split(col, ",")[0]            # first element
F.array_join(col, "-")          # array → string
```

---

## 🔸 9. Padding

```python
F.lpad(col, 10, "0")    # left pad
F.rpad(col, 10, "0")    # right pad
```

---

## 🔸 10. Repeat & Reverse

```python
F.repeat(col, 3)        # repeat string
F.reverse(col)          # reverse string
```

---

## 🔸 11. Null Handling

```python
F.coalesce(col, F.lit("default"))
F.na.fill({"col": "default"})
```

---

## 🔸 12. ASCII & Encoding

```python
F.ascii(col)        # ASCII value of first char
F.base64(col)       # encode
F.unbase64(col)     # decode
```

---

## 🔸 13. Conditional String Logic

```python
F.when(F.col("name").startswith("A"), "StartsWithA").otherwise("Other")
```

# 🔹 Pro Tips (Senior-Level)

* **Prefer built-in functions over UDFs** → Catalyst optimizer can optimize them.
* `substring` is **1-based index** (common mistake).
* Use `regexp_replace` instead of Python string ops → distributed & scalable.
* Chain transformations in `.select()` instead of multiple `.withColumn()` for better readability.
* For heavy pipelines → consider **column alias reuse** to avoid recomputation.
